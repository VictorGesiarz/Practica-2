```python
import re
import pandas as pd
```

# Merge authors generated data
Cargamos los datos generados de los autores, tenemos dos archivos diferentes que deberían tener los mismos resultados. Tenemos dos resultados porque lo ejecutamos dos veces, en la primera tuvimos errores con guardando los nuevos libros y tuvimos que volver a ejecutarla.

Cargamos los dos datos generadados.

En los datos generados por segunda vez, rellenamos los NAs en la columna **genres** por los valores del author1, con esto conseguimos eliminar algunos NAs.

Podriamos hacer lo mismo con la columna libros pero como los libros que tiene los primeros datos no tenemos su información guardada, no tendría sentido porque luego no los podríamos referenciar.

```python
input_folder = "../generated/"
output_folder = "../clean_data/"

authors1 = pd.read_csv(input_folder + "authors0.csv")
authors2 = pd.read_csv(input_folder + "authors2.csv")


print(authors1.isna().sum())
print(authors2.isna().sum())



authors1['Books'].replace("[]", pd.NA, inplace=True)


authors2['Genres'].fillna(authors1['Genres'], inplace=True)

print(authors1.isna().sum())
print(authors2.isna().sum())


authors2.to_csv(output_folder + "authors.csv", index=False)
```

# Merge books generated data

Tenemos tres archivos
- books2 que aprovechamos los datos del dataset Books que descargamos y le generamos nuevas columnas de información con la api de wikipedia.

- new_books1 libros que hemos añadido porque su autor estaba en nuestra base de datos.
- new_books2 es la continuación de new_books1 que hemos ejecutado después de que la de new_books1 se nos detuviera.

```python
books = pd.read_csv(input_folder + "books2.csv")

print(f"Books without rating: {sum(books['Rating Count'] == 0)}")

books
```

```python
new_books1 = pd.read_csv(input_folder + "new_books1.csv")
new_books1
```

```python
new_books2 = pd.read_csv(input_folder + "new_books2.csv")
new_books2
```

Concatenamos los dos datasets de new_books. Los nuevos libros no tienen rating, por eso, su columna de rating count debe ser 0. 

```python
new_books = pd.concat([new_books1, new_books2], ignore_index=True)
new_books['Rating Count'].fillna(0, inplace=True)
new_books
```

Eliminamos "(novel)" del título de los nuevos libros generados

```python
new_books['Book Title'] = new_books['Book Title'].apply(lambda x: re.sub(r'\([^)]*\)', '', x))
new_books
```

Eliminamos las columunas wikipedia id y free base id porque no aportan información.

Convertimos de variables flotantes a enteras number of pages, rating count y publication year.

```python
books = books.drop(['Wikipedia ID', 'Freebase ID'], axis=1)
books['Number of Pages'].fillna(pd.NA, inplace=True)
books['Number of Pages'] = books['Number of Pages'].astype(pd.Int64Dtype())
books['Rating Count'] = books['Rating Count'].astype(int)

books['Publication Year'].fillna(pd.NA, inplace=True)
books['Publication Year'] = books['Publication Year'].astype(pd.Int64Dtype())
books
```

Eliminamos los libros que ya están en books para que no estén en new_books. Para ello, miramos si el título y el autor coinciden.

```python
print(new_books.shape)
new_books = new_books[~new_books.apply(lambda x: (x['Book Title'] in books['Book Title'].values), axis=1)]
# new_books = new_books[~new_books.apply(lambda x: (x['Book Title'] in books['Book Title'].values) and (x['Author'] in books['Author'].values), axis=1)]
print(new_books.shape)
```

Concatenamos books y new books y los mezclamos para que los datos no estén ordenados. Podemos diferenciar los dos datasets porque los datos de books tendrán un rating count mayor a 0 a diferencia de los libros nuevos.

```python
books = pd.concat([books, new_books], ignore_index=True)

books = books.sample(frac=1).reset_index(drop=True)

books
```

```python
print(books.shape)
print(books['Book Title'].nunique())

repeated_titles = books[books.duplicated(subset='Book Title', keep=False)].sort_values(by='Book Title')
repeated_titles
```

```python
books.to_csv(output_folder + "books.csv", index=False)
```
