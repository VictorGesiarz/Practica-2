```python
import pandas as pd
import ast
```

```python
libros = pd.read_csv("./fiction_books.csv")
libros['Genres'] = libros['Genres'].apply(ast.literal_eval)
```

```python
libros_copia = libros.copy()
libros_copia = libros_copia[libros_copia["Genres"].apply(lambda x: len(x) >= 4)]
libros_copia.to_csv("./fiction_books_4genres.csv", index=False)
```
