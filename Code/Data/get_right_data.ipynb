```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv("./final_data/books.csv")
books = books.drop(books.columns[0], axis=1)
books
```

```python
books_copy = books.copy()
new_genres = []
for i, row in books_copy.iterrows():
    genres = row.loc["speculative":"comedy"]
    row_genres = []

    for genre, value in genres.items():
        if value:
            row_genres.append(genre)
    new_genres.append(row_genres)
books_copy["Genres"] = new_genres
```

```python
remove = books_copy.loc[:, "speculative":"comedy"]
books_copy = books_copy.drop(remove, axis=1)
books_copy
```

```python
books_copy.to_csv("./books.csv", index=False)
```

```python
books = books_copy.copy()
books = books.drop(columns=["Summary", "Followed by", "Preceded by"], axis=1)

ANTIQUITY = ["Recently published", "1 - 2 years", "3 - 10 years", "10 - 50 years", "More than 50 years"]
PAGES = ["0 - 100 pages", "100 - 300 pages", "300 - 600 pages", "More than 600 pages", "Unknown"]
YEAR = 2023


antiquity_list = []
pages_list = []
saga_list = []
for i, row in books.iterrows():
    antiquity = row["Publication Year"]
    if YEAR - antiquity > 50: antiquity = ANTIQUITY[4]
    elif YEAR - antiquity > 10: antiquity = ANTIQUITY[3]
    elif YEAR - antiquity > 3: antiquity = ANTIQUITY[2]
    elif YEAR - antiquity > 1: antiquity = ANTIQUITY[1]
    elif YEAR - antiquity > 0: antiquity = ANTIQUITY[0]
    antiquity_list.append(antiquity)

    pages = row["Number of Pages"]
    if 0 < pages <= 100: pages = PAGES[0]
    elif 100 < pages <= 300: pages = PAGES[1]
    elif 300 < pages <= 600: pages = PAGES[2]
    elif 600 < pages: pages = PAGES[3]
    else: pages = PAGES[4]
    pages_list.append(pages)

    saga = row["Saga"]
    if saga == "No": saga = "No"
    else: saga = "Yes"
    saga_list.append(saga)

books["Publication Year"] = antiquity_list
books["Number of Pages"] = pages_list
books["Saga"] = saga_list
```

```python
books['Evaluation count'] = 0
books['Evaluation mean'] = 0

# Add a new row with the first column equal to 0 and the rest equal to None
new_row = pd.Series([0] + [np.nan] * (books.shape[1] - 1), index=books.columns)

# Append the new row to the DataFrame
books = pd.concat([new_row.to_frame().T, books], ignore_index=True)
```
