```python
import pandas as pd
```

```python
books = pd.read_csv("./final_data/books.csv")
books = books.drop(books.columns[0], axis=1)
books
```

```python
books_copy = books.copy()
new_genres = []
for i, row in books_copy.iterrows():
    genres = row.loc["speculative":"comedy"]
    row_genres = []

    for genre, value in genres.items():
        if value:
            row_genres.append(genre)
    new_genres.append(row_genres)
books_copy["Genres"] = new_genres
```

```python
remove = books_copy.loc[:, "speculative":"comedy"]
books_copy = books_copy.drop(remove, axis=1)
books_copy
```

```python
books_copy.to_csv("./books.csv", index=False)
```
